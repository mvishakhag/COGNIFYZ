**Level 1 - Task 1 (Top Cuisines Analysis)**

```python
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd

# Get the filename dynamically
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

df.head()  # Check the first few rows
```

```python
# Splitting cuisines where multiple cuisines are listed together
cuisine_counts = df["Cuisines"].dropna().str.split(", ").explode().value_counts()

# Get the top 3 cuisines
top_3_cuisines = cuisine_counts.head(3)
print("Top 3 Cuisines:\n", top_3_cuisines)
```

```python
total_restaurants = len(df)
top_3_cuisines_percentage = (top_3_cuisines / total_restaurants) * 100
print("\nPercentage of Restaurants Serving Top 3 Cuisines:\n", top_3_cuisines_percentage)
```
